```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dataset = pd.read_csv('../Data/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree= 2)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Poliynomial regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
poly_reg2 = PolynomialFeatures(degree=4)
X_poly2 = poly_reg2.fit_transform(X)
lin_reg3 = LinearRegression()
lin_reg3.fit(X_poly2, y)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg3.predict(poly_reg2.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression - 4 degree)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
#linear regression
lin_reg.predict([[6.5]])
```

```python
#polinomial regression with 2 degree
lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```

```python
#polinomial regression with 4 degree
lin_reg3.predict(poly_reg2.fit_transform([[6.5]]))
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg3.predict(poly_reg2.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression - 4 degree)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
